```python
airplanedata = pd.read_csv('Airplane_Crashes.csv')
```

```python
airplanedata
```

```python
airplanedata['Year'] = pd.DatetimeIndex(airplanedata['Date']).year
```

```python
yearairtotals = pd.DataFrame(airplanedata.groupby(by = 'Year').count().Date)
```

```python
yearairtotals.reset_index(inplace = True)
yearairtotals
```

```python
airyearcrashes = yearairtotals.reset_index().values
```

```python
carcrashdata = pd.read_csv('out.csv')
```

```python
carcrashdata[['Fatalities', 'Injured persons', 'Crashes',
       'Vehicle-miles (millions)','Injured persons.1']] = carcrashdata[['Fatalities', 'Injured persons', 'Crashes',
       'Vehicle-miles (millions)','Injured persons.1']].replace({',':'','\(R\)':''},regex = True)
```

```python
carcrashdata.rename(columns = {'Rates per 100 million vehicle-miles':'Rates per 100m',
       'Fatalities.1':'Fatalities per 100m', 'Injured persons.1':'Injured persons per 100m', 'Crashes.1':'Crashes per 100m'},inplace = True)
```

```python
yearcartotals = carcrashdata[['Year', 'Crashes']]
```

```python
test = pd.merge(yearcartotals, yearairtotals,how='inner', on='Year')
```

```python
crashcompare = test[test.Year >= 1990].rename(columns = {'Year':'year','Crashes':'car_cr','Date':'air_cr'})
```

```python
crashcompare
```

```python
crashcompare.plot.line(x='year',y='air_cr')
```

```python
airfatal =airplanedata.groupby(by='Year').sum()['Fatalities'].reset_index()
```

```python
fatalities = pd.merge(airfatal, carcrashdata,how='inner', on='Year')
```

```python
fatalities.rename(columns = {'Fatalities_x':'air_fatalities','Fatalities_y':'car_fatalities'},inplace= True)
```
